```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.svm import SVR
```

```python
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent=data[['charges']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
kernel_p=['rbf','poly','sigmoid','linear']
C_p=[10,100,1000,2000,3000]
gamma_p=['auto','scale']
k_array=[]
c_array=[]
g_array=[]
r_array=[]
for i in kernel_p:
    for j in C_p:
        for k in gamma_p:
            regressor=SVR(kernel=i,C=j,gamma=k)
            regressor.fit(x_train,y_train)
            y_predict=regressor.predict(x_test)
            r_score=r2_score(y_test,y_predict)
            r_array.append(r_score)
            k_array.append(i)
            c_array.append(j)
            g_array.append(k)           
```

```python
dict={"kernel":k_array,"C":c_array,"gamma":g_array,"r2_score":r_array}
final_parameter=pd.DataFrame(dict)
final_parameter.to_csv("SVM_R2score.csv")
final_parameter
```

```python
regressor=SVR(C=3000,kernel='rbf',gamma='auto')
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
```

```python
r_score=r2_score(y_test,y_predict)
r_score
```
