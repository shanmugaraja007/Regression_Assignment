```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn import tree
```

```python
data=pd.read_csv("insurance_pre.csv")
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=data[['charges']]
```

```python
sc1=StandardScaler()
sc2=StandardScaler()
```

```python
x_train=sc1.fit_transform(independent)
y_train=sc2.fit_transform(dependent)
```

```python
param_grid={"criterion":['mse','mae','friedman_mse'],"splitter":['best','random'],'max_features':['auto','sqrt','log2']}
```

```python
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,n_jobs=-1,verbose=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
dic=grid.cv_results_
pd.DataFrame(dic)
```

```python
y_predict=grid.predict(x_train)
```

```python
r2=r2_score(y_train,y_predict)
r2
```
