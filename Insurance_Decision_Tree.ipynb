```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent=data[['charges']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
criterions=['mse','mae','friedman_mse']
splitters=['random', 'best']
max_feature=['sqrt', 'auto', 'log2']
a=[]
b=[]
c=[]
r2=[]
for x in criterions:
    for y in splitters:
        for z in max_feature:
                    regressor=DecisionTreeRegressor(criterion=x,splitter=y,max_features=z)
                    regressor.fit(x_train,y_train)
                    y_predict=regressor.predict(x_test)
                    r_score=r2_score(y_test,y_predict)
                    r2.append(r_score)
                    a.append(x)
                    b.append(y)
                    c.append(z)
```

```python
Dict={"criterion":a,"splitter":b,"max_features":c,"r2_scor":r2}
final_data=pd.DataFrame(Dict)
final_data.to_csv("Decision_tree_parameters.csv")
final_data
```

```python
regressor=DecisionTreeRegressor(criterion='mae',splitter='random',max_features='log2')
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
```

```python
r_score=r2_score(y_test,y_predict)
r_score
```
